# Feature hacking

This is a little R tutiorial showing how feature hacking, or selecting your features based on test set performance, can dramatically inflate model performance.

### Step 2: Generate Data

We will define a parameter object _p_ that has fields defining the data we will be generating. We'll simulated data for two groups (Group A, Group B) and the goal of our classifier is to use the features we generate to classify whether participants belong to Group A or Group B. 

We will generate responses from 30 independent brain regions (1 response per region to keep it simple) where half of the responses are drawn from different distributions for the two groups and the other half are drawn from the same distribution.

```r
# Sample sizes
p.n_sub = 34*2
p.n_group = 2
p.n_voxels = 1
p.n_regions = 30

# sampling parameters (all gaussian)
p.mu_a = runif(p.n_regions, min=-.7, max=.7) # mean of generative distirubtion for each region in group A
p.mu_b = runif(p.n_regions/2, min=-.7, max=.7) # mean of generative distirubtion for each region in group B
p.mu_b = c(p.mu_b, p.mu_a[(p.n_regions/2+1):p.n_regions]) # Let's make 1/2 of the regions come from different distributions
p.sigma_a = rep(1,p.n_regions)   # std of generative distirubtion for each region in group A
p.sigma_b = rep(1,p.n_regions)   # std of generative distirubtion for each region in group A
p.sigma_sample = 0.3 # Subject noise (assume it's equal across subjects)
```

Now the data generation function looks like this. It's also in this repository as **generate_data.R**.

```r
generate_data <- function(p){
  
  # Keep data object
  #reg_resp = matrix(NaN,nrow=p.n_sub,ncol=p.n_voxels*p.n_regions)
  full_resp = vector()
    for (r in 1:p.n_regions){
        reg_resp = matrix(NaN,nrow=p.n_sub,ncol=p.n_voxels) 
        for (s in 1:p.n_sub){
            if (s <= p.n_sub/p.n_group) {
                reg_resp[s,]= rnorm(p.n_voxels,mean=p.mu_a[r],sd=p.sigma_a[r])+rnorm(p.n_voxels,mean=0,sd=p.sigma_sample)  
              } else {
                if (r==p.n_regions) {
                    r
            }
                reg_resp[s,]= rnorm(p.n_voxels,mean=p.mu_b[r],sd=p.sigma_b[r])+rnorm(p.n_voxels,mean=0,sd=p.sigma_sample)  
            }
        }
   
  # Concatenate the data vector 
  full_resp = cbind(full_resp, reg_resp)
  }
  
  return(full_resp)
}
```

Let's run it and see what the data looks like on one sim.

```r
# Generate data file
reg_resp = generate_data(p)

# Show the data for now
plot(reg_resp,col=brewer.pal(n=10,name="RdBu"))
```

To show that this data set is decodable, we'll run a set of tests where we will use leave one out cross validation (LOOCV), with a naive Bayes classifier, to predict group. The function to do this looks like this.

```r
loocv_nb <- function(data, id){
  
  # Generate accuracy output
  n_sub = dim(data)[1]
  acc = matrix(NaN,nrow=n_sub,ncol=1)
  
  for (s in 1:n_sub){
    
    # Get your training & test subjects
    test_sub = s
    train_subs = setdiff(1:n_sub,s)
    
    # Use a naive bayes classifier
    nb = naive_bayes(x=data[train_subs,], y=id[train_subs])
    test = predict(nb, newdata=t(data.frame(data[test_sub,])))
    
    # Save the accuracy of the fold
    acc[s]=(test==id[test_sub]) 
  }
  
  # Push the output 
  return(acc)
}
```

Let's run 100 iterations generating data and measuring it's accuracy with the LOOCV test.

```r
# Make labels for groups
labels = matrix(c(rep('a',p.n_sub/p.n_group),rep('b',p.n_sub/p.n_group)),nrow=p.n_sub,ncol=1)

# Now let's loop through and see how well we do across iterations.
# this step takes a while
n_iterations = 100
acc = matrix(NaN,ncol=n_iterations,nrow=1)
for (i in 1:n_iterations){
    # Generate new data set
    reg_resp = generate_data(p)
    
    # Run LOOCV with Bayes classifier
    run_acc = loocv_nb(reg_resp, labels)
    
    # Store the accuracy
    acc[i]=mean(run_acc)
}

hist(acc)
```

Giving us a mean classification accuracy of:

```r
print(mean(acc))
```

This proves that we have a data set that can classify group, but it is not perfect. Which is good as it gives us range to work with.

### Step 3: Proper feature selection

The proper way to do feature selection is to only use a hold out training set. For this simulation, we will be using LASSO to do feature selection. Thus we actually need two hold out training sets. One to choose the sparsity parameter $\lambda$ and one to learn the LASSO model using that learned (or tuned) $\lambda$ value.

So we will generate up a data set that can be decomposed into three groups. The first two will be used to apply LASSO and find the features to be used in the final 


```r
# Now let's just generate a normative example (repeat until decent accuracy for 
# demonstration purposes)
p.n_sub = 34*3 # three groups of N=34 each
reg_resp = generate_data(p)
labels = matrix(c(rep('a',p.n_sub/p.n_group),rep('b',p.n_sub/p.n_group)),nrow=p.n_sub,ncol=1)
run_acc = mean(loocv_nb(reg_resp, labels))
print(run_acc)

# 3 groups: 1 to learn lambda, 1 to select regions/fit, 1 to test fit
subset_indx = rep(1:3,p.n_sub/3)

# Lists we are going to use to subset the data: only groups 1 & 2
training_subset = subset_indx %in% c(1,2)
```

So this is good, we can recover basically the same level of decoding accuracy we had before (which isn't surprising since this is a larger data set)

Now, I'll setup a function to use the first two groups in the set to find the optimal $\lambda$ and use it to find the regions to evaluate in the test set. The function looks like this.

```r
lasso_regionsearch <- function(data, id, subset){
  
  train_set = subset==1
  test_set  = subset==2
  eval_set  = subset==3
  
  # Find the best lasso constraint
  model.tune = cv.glmnet(data[train_set,], id[train_set], family="binomial")
  
  # Return lambda
  model.fit = glmnet(data[test_set,], id[test_set], family="binomial", lambda=model.tune$lambda.min)
  
  return(model.fit)
}
```

Since we are using logistic LASSO, we will need to redo the Y labels to be binary 0 & 1, instead of "a" & "b" labels.

```r
# Setup labels
region_indx = rep(1:p.n_regions, each=p.n_voxels)
bin_labels = matrix(c(rep(1,p.n_sub/p.n_group),rep(0,p.n_sub/p.n_group)),nrow=p.n_sub,ncol=1)

# Let's use LASSO to find the best regions
model = lasso_regionsearch(reg_resp,bin_labels, subset_indx)
best_regions = which(coef(model)!=0)
print(best_regions)
```

Now let's take the regions selected by LASSO and run a LOOCV in the hold out set.

```r
# Specify the test set
test_X = reg_resp[subset_indx==3,]
test_Y = labels[subset_indx==3]

# Take the output from LASSO and test in LOOCV
best_indx = region_indx %in% sort(best_regions)
correct_cv_acc = mean(loocv_nb(test_X[,best_indx],test_Y))
print(correct_cv_acc)
```

We can compare this accuracy against the LOOCV accuracy you would get if you left all the features in the model.

```r

# Compare against test set accuracy without dim reduction
test_acc = mean(loocv_nb(test_X, test_Y))
print(test_acc)
```

Most of the time, your reduced feature set will do worse because it is a lower dimensional model (i.e., it has less flexibility).

### Step 4: Feature hacking

Feature hacking consists of using your _test set_ to do your feature hacking. There are many ways that this can be done. Here I will present the **worst case** scenario as a demonstration.

Imagine that we really wanted to cheat, so that we kept searching through our variables on the test data set until we found a fit that "looked good". An brute force random search to do that would look like this (inspired by this paper https://arxiv.org/pdf/1803.07055.pdf).

```r
cheat_codes <- function(data, id, region_indx){
  
  regions = unique(region_indx)
  final_acc = vector()    # final output of acc values
  num_regions = vector()
  isDone = 0               # termination flag
  k = 1                    # counter for model complexity
  
  # Loop until you get a termination flag
  while (!isDone){
    
    # Brute force the run
    runs = permutation_run(k+1, data, id, region_indx)
    
    # find the best accuracy
    final_acc[k] = max(runs$acc)
    num_regions[k] = k+1

    # save the output
    if (k>1) {
      if (k==length(regions) | final_acc[k-1]>final_acc[k]){
        isDone = 1
        best_set = runs[runs[,2]==final_acc[k],]
      }
    }
    k=k+1
  }

  best_run = runs
  out = data.frame(final_acc[1:k-1], num_regions[1:k-1])
  return(out)
  
}
```

The function above relies on this permutation step, which looks through the data randomly selecting _k_ features, evaluating the performance then looking at performance if you increase the number fo features. It keeps searching until you get through all $p$ features (30 in our case) or if the accuracy on the $k^{th}$ run is more than the accuracy on the $k^{th}+1$

```r
permutation_run <- function(r, data, id, region_indx){
  
  regions = unique(region_indx)
  
  # Critical numbers
  n_iterations = 100
  n_regions = length(regions)
  
  # Output data objects
  regions_used = matrix(NaN,nrow=n_iterations, ncol=r)
  acc          = matrix(NaN,nrow=n_iterations, ncol=1)
  
  # Loop through random permutations
  for (i in 1:n_iterations){
    
    # New subsets of words & regions
    regions2use = sample(1:n_regions)[1:r]
    
    indx = region_indx %in% sort(regions2use)
    new_data = data[,indx]
      
    if (length(regions2use < 2)) {
      new_data = t(t(new_data));
    }
    
    acc[i]= mean(loocv_nb(new_data, id))
    regions_used[i,] = regions2use
  }
  
  out = data.frame(regions_used, acc)
  return(out)
  
}
```

This is about as academically dishonest as you can get. You're basically fishing for the best combination of features in your data set that give you your best results. 

Here's how it looks in practice.

```r
# Now let's REALLY overfit. I mean just plain cheat
cheat = cheat_codes(test_X, test_Y, region_indx)
n_regions = dim(cheat)[1]-1
```

Now that we found our best feature set size, we can re-run our permutation test to find the best combination of $k$ features.

```r
re_run_boot = permutation_run(n_regions, test_X, test_Y, region_indx)
```

We will now select the best model out of this set and call that our best performance.

```r

# Take the best model and call it your best
best_model = re_run_boot[re_run_boot$acc == max(re_run_boot$acc),]
if (dim(best_model)[1]>1){
  best_model = best_model[1,]
}

cheat_acc = best_model$acc
print(cheat_acc)
```

Notice that this is _substantially_ better than what we saw with our more principled feature selection approaches. That's by design.

### Step 5: Test for feature hacking

If you used a principled method to select your features independent of the hold out test set, then you will miss some meaningful variance in the test set. That's because it is, by nature, a conservative approach. 

**However, the more you overfit, the more unique variance in your test set you are able to capture. Which means, the probability of finding a combination of features that gives you better peformance is very, very low.**

Let's leverae our random search approach to help us test for overfitting. We know that the **cheat_acc** result is due to overfitting by using the test set. What we can do is re-search the test set for different combinations of features and ask how often we see a fit that is **better**. 

Using a conservative hold out test to do feature selection (i.e., not overfitting), however, should result in finding combinations of features in the test set that yield better peformance.

The **sim_cheat_exp_with_lasso.R** function does this search. On each iteration it compares the probability of finding a better model, via random search, than the model with the features selected from the independent hold out test set (**test_acc** above) or than the overfit model performance (**cheat_acc**).

(Note: This takes a while. So if you are running the code here, go to bed and come back).


```r
perm_test = sim_cheat_exp_with_lasso(p, 100)
```

Across all searches the probability of finding a better model than the appropriate feature selection method is:

```r
print(mean(perm_test$p_cv_greater))
```

The probability of finding a better model than our overfit model is:

```r
print(mean(perm_test$p_cheat_greater))
```

# Summary

1. It is possible to over-fit with cross-validated accuracies by using the test set to select your features. I call this feature hacking.

2. A signature of this effect would be to compare the observed results against a random feature search and calculate the probability that you'd find a better combination of features from random chance. The smaller this probability, the more likely the outcome measure reflects overfitting.
